```python
d = 5
def construct_symmetric_matrix(Theta):
    Theta_matrix = np.zeros((d , d))
    upper_idx = np.triu_indices(d)
    Theta_matrix[upper_idx] = Theta
    Theta_matrix = Theta_matrix + Theta_matrix.T - np.diag(np.diag(Theta_matrix)) 
    return(Theta_matrix)

def pen_log_likelihood(Theta , EstimSigma , Lambda) :
    Theta_matrix  = construct_symmetric_matrix(Theta)
    abs_val_Theta_matrix = np.absolute(Theta_matrix)
    upper_idx = np.triu_indices(d)
    upper_triangle_abs_val_Theta_matrix =  abs_val_Theta_matrix[upper_idx]
    off_diag_sum = upper_triangle_abs_val_Theta_matrix.sum() - abs_val_Theta_matrix.trace()
    print("This is is the off diagonal sum:", off_diag_sum)
    det_Theta = np.linalg.det(Theta_matrix) 
    if det_Theta <= 0 :
        return -np.inf
    product_STheta = np.matmul(EstimSigma, Theta_matrix)
    trace_STheta =  product_STheta.trace()
    return(np.log(det_Theta) - trace_STheta - Lambda * off_diag_sum )

    
#A = np.random.randn(d, d)
#nitial_Theta_matrix = np.dot(A, A.T) + d * np.eye(d) 

# Convert to upper-triangular vector form
#upper_idx = np.triu_indices(d)
#initial_guess = initial_Theta_matrix [upper_idx]
#Lambda = 10
#random.seed(7)
#result = optimize.minimize(lambda x: -log_likelihood(x , emp_cov , Lambda) , np.array(initial_guess) ).x
#mle =  construct_symmetric_matrix(result)
#print("The maximum likelihood estimator of the precision matrix is given by \n", mle)    
```
