This notebook demonstrates how to compute covariance, precision, correlation, and scaled inverse correlation matrices for the math grades dataset.

```python
import pandas
import numpy as np

# Load the dataset (adjust path as needed)
df = pandas.read_csv('data/mathmarks.csv')

# Inspect the shape of the dataset
print(df)

```

## Convert to NumPy and Compute Covariance Matrix

```python
# Convert to NumPy and transpose
matrix = df.to_numpy().T

# Covariance matrix
cov_mat = np.cov(matrix)
print(cov_mat)
```

## Precision Matrix

```python
precision_mat = np.linalg.inv(cov_mat) * 1000
print(precision_mat)
```
