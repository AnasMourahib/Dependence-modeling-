```python
import numpy as np 
import scipy.stats
import pandas
import scipy.optimize as optimize
import random
from sklearn.datasets import make_sparse_spd_matrix
from sklearn.covariance import empirical_covariance, graphical_lasso


csvFile = pandas.read_csv('C:/Users/20254817/Desktop/Githib/Dependence-modeling-/data/student_exam_scores.csv')



data = csvFile.to_numpy()
data = data[: , np.arange(1,6)]
emp_cov = empirical_covariance(data)
#print(emp_cov)

precision_mat = np.linalg.inv(emp_cov)
#print(precision_mat)

cov_, precision_, costs_ , n_iter_ = graphical_lasso(emp_cov, alpha=10 , return_costs = True , return_n_iter = True)

print("Estimated covariance matrix:\n", cov_)
print("\nEstimated precision matrix:\n", precision_)
print("\nEstimated cost:\n", costs_)
print("\nnumber of iterations:\n", n_iter_)

upper_idx = np.triu_indices(5)
theta_matrix =  precision_[upper_idx]
print(theta_matrix)

cost = log_likelihood(theta_matrix , emp_cov , 10)
print(cost)
```

```python
d = 5
def construct_symmetric_matrix(Theta):
    Theta_matrix = np.zeros((5 , 5))
    upper_idx = np.triu_indices(5)
    Theta_matrix[upper_idx] = Theta
    Theta_matrix = Theta_matrix + Theta_matrix.T - np.diag(np.diag(Theta_matrix)) 
    return(Theta_matrix)

def log_likelihood(Theta , EstimSigma , Lambda) :
    Theta_matrix  = construct_symmetric_matrix(Theta)
    print(Theta_matrix)
    det_Theta = np.linalg.det(Theta_matrix) 
    if det_Theta <= 0 :
        return -np.inf
    product_STheta = np.matmul(EstimSigma, Theta_matrix)
    trace_STheta =  product_STheta.trace()
    return(np.log(det_Theta) - trace_STheta - Lambda * Theta_matrix.sum() )

    
A = np.random.randn(d, d)
initial_Theta_matrix = np.dot(A, A.T) + d * np.eye(d) 

# Convert to upper-triangular vector form
upper_idx = np.triu_indices(d)
initial_guess = initial_Theta_matrix [upper_idx]
Lambda = 10
random.seed(7)
result = optimize.minimize(lambda x: -log_likelihood(x , emp_cov , Lambda) , np.array(initial_guess) ).x
mle =  construct_symmetric_matrix(result)
print("The maximum likelihood estimator of the precision matrix is given by \n", mle)    
```
